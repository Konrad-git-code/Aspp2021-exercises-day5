## Statistics

### a.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
disc = stats.poisson(3.5)
n = np.arange(0, 20)
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].plot(n, disc.pmf(n))
axes[1].plot(n, disc.cdf(n))
axes[2].hist(disc.rvs(size=1000))
fig.subplots_adjust(wspace = 0.5)
```

## b.

```python
cont = stats.norm()
n = np.linspace(-6, 6, 1000)
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].plot(n, cont.pdf(n))
axes[1].plot(n, cont.cdf(n))
axes[2].hist(cont.rvs(size=1000))
fig.subplots_adjust(wspace = 0.5)
```

### c.

```python
stats.ttest_ind(disc.rvs(size=1000), disc.rvs(size=1000))
```

```python
stats.ttest_ind(cont.rvs(size=1000), cont.rvs(size=1000))
```
